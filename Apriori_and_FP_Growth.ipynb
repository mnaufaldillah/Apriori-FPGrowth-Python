# Apriori

```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori

apriori(df, min_support=0.5)
```

```python
apriori(df, min_support=0.5, use_colnames=True)
```

```python
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets[ (frequent_itemsets['length'] >= 2) &
                   (frequent_itemsets['support'] >= 0.8) ]
```

```python
frequent_itemsets[ frequent_itemsets['itemsets'] == frozenset(('Eggs', 'Onion')) ]
```

```python
oht_ary = te.fit(dataset).transform(dataset, sparse=True)
sparse_df = pd.DataFrame.sparse.from_spmatrix(oht_ary, columns=te.columns_)
sparse_df
```

```python
apriori(sparse_df, min_support=0.6, use_colnames=True)
```

```python
from google.colab import files
from pyspark.sql import functions as F
from pyspark.ml.fpm import FPGrowth
import pandas


spark_data = spark.createDataFrame(data)
basket_data = spark_data.dropDuplicates(['SalesTransactionID', 'SalesItem']).sort('SalesTransactionID')
basket_data = basket_data.groupBy("SalesTransactionID").agg(F.collect_list("SalesItem")).sort('SalesTransactionID')
```

# FP Growth - metode mining itemsets yang sering muncul

```python
fp_Growth = FPGrowth(itemsCol="collect_list(SalesItem)", minSupport=0.005, minConfidence=0.005) 
model = fp_Growth.fit(basket_data)
```

# Menampilkan Itemsets yang sering muncul

```python
model.freqItemsets.show()
items = model.freqItemsets
```

# Menampilkan Association Rules yang dibuat

```python
model.associationRules.show()
rules = model.associationRules
```

# transform memeriksa input item terhadap semua association rule dan merangkum konsekuen sebagai prediksi

```python
model.transform(basket_data).show()
transformed = model.transform(basket_data)
```
